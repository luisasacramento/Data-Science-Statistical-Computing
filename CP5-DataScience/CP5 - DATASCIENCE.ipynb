```python
#1-C
def winsorize(df):
    if not isinstance(df, pd.DataFrame):
        raise ValueError("O objeto passado não é um DataFrame")
    
    df = df.fillna('others')  

    for col in df.select_dtypes(include=['number']).columns:
        media = df[col].mean()
        df[col].fillna(media, inplace=True)

    for col in df.select_dtypes(include=['number']).columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR
        
     
        df[col] = df[col].clip(lower=limite_inferior, upper=limite_superior)

        

    return df


df2 = winsorize(df)
print(df2)

for col in df2.select_dtypes(include=['number']).columns:
    media = df[col].mean()
    count_others_replaced = (df2[col] == media).sum()
    print(f"Número de vezes que a média ({media}) aparece na coluna '{col}': {count_others_replaced}")
```
